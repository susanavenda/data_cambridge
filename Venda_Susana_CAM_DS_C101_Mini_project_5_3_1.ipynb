```python
# Create binary flags for outliers
outlier_df = pd.DataFrame()
for col in features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outlier_df[f'{col}_outlier'] = ((df[col] < (Q1 - 1.5*IQR)) | (df[col] > (Q3 + 1.5*IQR))).astype(int)

# Flag if 2 or more features show extreme behavior
df['iqr_anomaly'] = (outlier_df.sum(axis=1) >= 2).astype(int)
```

Standardisation (Z-score scaling) is the correct choice. Here is why:

Outlier Sensitivity: Your primary goal is anomaly detection. Normalisation squashes all data into a 0 to 1 range. If you have a massive outlier (like a 500°C temperature spike), normalisation will cram all your "normal" data into a tiny range (e.g., 0.001 to 0.002), making it invisible to the model. Standardisation handles outliers much better by centering them around the mean.

Algorithm Requirements: * SVM: Relies on the distance between points. It performs significantly better when features are standardized.

K-Means: Uses Euclidean distance. If RPM (0–3000) isn't standardized against Pressure (0–5), the RPM will dominate the calculation entirely.

PCA: You used PCA for your plots. PCA requires standardisation to correctly identify the directions of maximum variance.

We selected Standardisation over Normalisation because our dataset contains critical outliers (anomalies). Standardisation ensures that these anomalies maintain their relative distance from the mean, preventing the 'data squashing' effect that occurs with Min-Max scaling, thereby improving the accuracy of our distance-based models (SVM and K-Means).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. Load and Preprocess
url = "https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/engine.csv"
df = pd.read_csv(url)

# Use the features we engineered previously
df['lub_efficiency'] = df['Lub oil pressure'] / (df['lub oil temp'] + 1e-6)
df['cooling_delta'] = df['Coolant pressure'] - df['Coolant temp']
features = ['Engine rpm', 'Lub oil pressure', 'Fuel pressure',
            'Coolant pressure', 'lub oil temp', 'Coolant temp',
            'lub_efficiency', 'cooling_delta']

# 2. Scaling & PCA (Reducing to 2D for the visualization grid)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 3. Fit One-Class SVM
# nu=0.03 aims for the 3% anomaly target
model = OneClassSVM(kernel='rbf', gamma=0.1, nu=0.03)
model.fit(X_pca)
df['svm_anomaly'] = model.predict(X_pca)

# 4. Create Visualization Grid
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

# Calculate decision function across the grid
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 5. Plotting
plt.figure(figsize=(12, 8))
plt.title("One-Class SVM: Engine Operational 'Safety Zone'", fontsize=15)

# Plot the learned "Normal" boundary
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap=plt.cm.PuBu, alpha=0.8)
plt.contour(xx, yy, Z, levels=[0], linewidths=3, colors='darkred') # The Boundary
plt.contourf(xx, yy, Z, levels=[0, Z.max()], colors='palevioletred', alpha=0.4)


# Plot the actual engine data points
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1],
                hue=np.where(df['svm_anomaly'] == 1, 'Normal', 'Anomaly'),
                palette={'Normal': 'deepskyblue', 'Anomaly': 'red'},
                style=np.where(df['svm_anomaly'] == 1, 'Normal', 'Anomaly'),
                markers={'Normal': 'o', 'Anomaly': 'X'}, s=60)



# Plot the actual engine data points
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1],
                hue=np.where(df['svm_anomaly'] == 1, 'Normal', 'Anomaly'),
                palette={'Normal': 'deepskyblue', 'Anomaly': 'red'},
                style=np.where(df['svm_anomaly'] == 1, 'Normal', 'Anomaly'),
                markers={'Normal': 'o', 'Anomaly': 'X'}, s=60)
plt.xlabel('Principal Component 1 (Engine Load/Speed)')
plt.ylabel('Principal Component 2 (Thermal/Pressure Efficiency)')
plt.legend(title="Engine State")
plt.show()

# Print results
anomaly_count = len(df[df['svm_anomaly'] == -1])
print(f"Total Anomalies Detected: {anomaly_count} ({anomaly_count/len(df)*100:.2f}%)")
```
