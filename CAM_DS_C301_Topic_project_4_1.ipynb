```python
## @title 0.2 Imports, palette and display helpers
import re
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
from wordcloud import WordCloud

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist

from IPython.display import display, HTML

warnings.filterwarnings('ignore')
pio.renderers.default = "colab"

# ── Palette & theme (same as CAM_C201 mini-project) ───────────────────────
DARK_BG   = "#0F172A"
CARD_BG   = "#1E293B"
BORDER    = "#334155"
TEXT_PRI  = "#F1F5F9"
TEXT_SEC  = "#94A3B8"
TEAL      = "#0EA5E9"
GREEN     = "#22C55E"
AMBER     = "#F59E0B"
RED       = "#EF4444"
PURPLE    = "#A78BFA"

PALETTE   = [TEAL, GREEN, AMBER, RED, PURPLE, "#FB923C", "#34D399"]
STOP_WORDS = set(stopwords.words('english'))

# ── Display helpers ────────────────────────────────────────────────────────
_fig_counter, _table_counter = [0], [0]

def _show_fig(fig, title=""):
    _fig_counter[0] += 1
    label = f"Figure {_fig_counter[0]}" + (f" — {title}" if title else "")
    fig.update_layout(
        plot_bgcolor=CARD_BG, paper_bgcolor=DARK_BG,
        font_color=TEXT_PRI, font_family="Inter, DejaVu Sans",
        title_font_color=TEXT_PRI,
        xaxis=dict(gridcolor=BORDER, linecolor=BORDER),
        yaxis=dict(gridcolor=BORDER, linecolor=BORDER),
        margin=dict(l=40, r=20, t=50, b=40),
    )
    display(HTML(f'<p style="font-family:Inter,sans-serif;font-size:11px;color:{TEXT_SEC};margin:8px 0 2px">{label}</p>'))
    display(HTML(fig.to_html(full_html=False, include_plotlyjs="cdn")))

def display_table(df, title="", max_rows=30):
    _table_counter[0] += 1
    label = f"Table {_table_counter[0]}" + (f" — {title}" if title else "")
    cols = list(df.columns)
    header = "".join(f'<th style="padding:8px 14px;text-align:left;color:{TEAL};font-weight:600;'
                      f'border-bottom:1px solid {BORDER};font-family:Inter,sans-serif;font-size:12px">{c}</th>' for c in cols)
    rows_html = ""
    for i, (_, row) in enumerate(df.head(max_rows).iterrows()):
        bg = CARD_BG if i % 2 == 0 else "#253047"
        cells_html = "".join(f'<td style="padding:7px 14px;font-family:Inter,sans-serif;font-size:12px;'
                              f'color:{TEXT_PRI}">{v}</td>' for v in row.values)
        rows_html += f'<tr style="background:{bg}">{cells_html}</tr>'
    display(HTML(f'''
        <p style="font-family:Inter,sans-serif;font-size:11px;color:{TEXT_SEC};margin:12px 0 4px">{label}</p>
        <div style="overflow-x:auto;border-radius:8px;border:1px solid {BORDER}">
          <table style="border-collapse:collapse;width:100%;background:{DARK_BG}">
            <thead><tr>{header}</tr></thead><tbody>{rows_html}</tbody>
          </table>
        </div>'''))

def insight(text, colour=TEAL):
    display(HTML(f'''
        <div style="border-left:3px solid {colour};background:{CARD_BG};padding:12px 16px;
                    border-radius:0 6px 6px 0;margin:12px 0">
          <p style="margin:0;font-family:Inter,sans-serif;font-size:13px;color:{TEXT_PRI};line-height:1.7">{text}</p>
        </div>'''))

def section_header(title, subtitle="", colour=TEAL):
    sub_html = (f'<p style="margin:6px 0 0;font-size:13px;color:{TEXT_SEC};font-family:Inter,sans-serif">{subtitle}</p>'
                if subtitle else "")
    display(HTML(f'''
        <div style="border-bottom:2px solid {colour};padding-bottom:10px;margin:28px 0 18px">
          <h2 style="margin:0;font-size:18px;font-weight:700;color:{TEXT_PRI};font-family:Inter,sans-serif">{title}</h2>
          {sub_html}
        </div>'''))

def kpi_row(items):
    """items = list of (label, value, delta_text, positive_bool)"""
    cards = ""
    for label, value, delta, pos in items:
        dc = GREEN if pos else RED
        cards += f'''
          <div style="background:{CARD_BG};border:1px solid {BORDER};border-radius:8px;padding:16px 20px;
                      flex:1;min-width:140px">
            <p style="margin:0 0 4px;font-size:11px;color:{TEXT_SEC};font-family:Inter,sans-serif;font-weight:600;
                      text-transform:uppercase;letter-spacing:.05em">{label}</p>
            <p style="margin:0;font-size:24px;font-weight:700;color:{TEXT_PRI};font-family:Inter,sans-serif">{value}</p>
            <p style="margin:4px 0 0;font-size:11px;color:{dc};font-family:Inter,sans-serif">{delta}</p>
          </div>'''
    display(HTML(f'<div style="display:flex;gap:12px;flex-wrap:wrap;margin:12px 0">{cards}</div>'))

def plot_wordcloud(text_blob, title=""):
    wc = WordCloud(width=900, height=400, background_color=DARK_BG,
                   colormap="GnBu", max_words=100).generate(text_blob)
    fig, ax = plt.subplots(figsize=(10, 4.4), facecolor=DARK_BG)
    ax.imshow(wc, interpolation="bilinear")
    ax.axis("off")
    ax.set_title(title, color=TEXT_PRI, fontsize=13, family="DejaVu Sans")
    plt.show()

def plot_top_words(freq_dist, n=10, title="", colour=TEAL):
    top = freq_dist.most_common(n)
    words, counts = zip(*top)
    fig = go.Figure(go.Bar(x=list(counts)[::-1], y=list(words)[::-1], orientation="h", marker_color=colour))
    fig.update_layout(title=title, height=380, xaxis_title="Frequency")
    _show_fig(fig, title)

print("✓ Setup complete")
```

```python
## @title 1.1 Load the two review data sets into SQLite
section_header("Loading the data", "Google and Trustpilot, 12 months of PureGym reviews — via SQLite")

!pip install gdown openpyxl --quiet
import gdown
import sqlite3
import pandas as pd
import os

# ── Download from Drive ─────────────────────────────────────────────────
DOWNLOAD_DIR = '/tmp/project_data'
os.makedirs(DOWNLOAD_DIR, exist_ok=True)

GOOGLE_PATH     = os.path.join(DOWNLOAD_DIR, 'Google_12_months.xlsx')
TRUSTPILOT_PATH = os.path.join(DOWNLOAD_DIR, 'Trustpilot_12_months.xlsx')

files_to_download = {
    GOOGLE_PATH:     "13TeZTT6oqk3iPELBhFx9hComIfChKw4m",
    TRUSTPILOT_PATH: "1V6Pr6xWrHSHaothB61MJbR7mPFHwQ1u6",
}
for filepath, file_id in files_to_download.items():
    url = f"https://drive.google.com/uc?id={file_id}"
    gdown.download(url, filepath, quiet=False)

# ── Load into pandas ────────────────────────────────────────────────────
google_raw     = pd.read_excel(GOOGLE_PATH)
trustpilot_raw = pd.read_excel(TRUSTPILOT_PATH)

# ── Cast oversized/mixed-type ID columns to strings (avoids SQLite OverflowError) ──
def safe_for_sqlite(df):
    df = df.copy()
    for col in df.columns:
        if df[col].dtype == object:
            df[col] = df[col].apply(lambda x: x if pd.isna(x) else str(x))
    return df

google_raw_safe     = safe_for_sqlite(google_raw)
trustpilot_raw_safe = safe_for_sqlite(trustpilot_raw)

# ── Load into SQLite ─────────────────────────────────────────────────────
conn = sqlite3.connect("/tmp/puregym_reviews.db")
google_raw_safe.to_sql("google_reviews_raw", conn, if_exists="replace", index=False)
trustpilot_raw_safe.to_sql("trustpilot_reviews_raw", conn, if_exists="replace", index=False)

kpi_row([
    ("Google rows (DB)",     f"{len(google_raw):,}",     "loaded into google_reviews_raw", True),
    ("Trustpilot rows (DB)", f"{len(trustpilot_raw):,}", "loaded into trustpilot_reviews_raw", True),
])

insight("Both files were sourced from a shared Google Drive folder rather than direct upload, so they're pulled "
        "in via gdown using each file's Drive file ID. Data is loaded into a local SQLite database "
        "(puregym_reviews.db) rather than kept purely in pandas — this gives a permanent, queryable copy of the "
        "raw, untouched data (google_reviews_raw, trustpilot_reviews_raw) that can always be checked against "
        "later, even after the working dataframes have been cleaned and transformed many steps downstream. "
        "Implementation note: writing Trustpilot's raw data to SQLite initially failed with "
        "'OverflowError: Python int too large to convert to SQLite INTEGER', caused by an ID-like column "
        "(e.g. Business Unit ID) storing a number that exceeds SQLite's 64-bit integer limit. Fixed by casting "
        "all object-dtype columns to plain strings before writing, while leaving true NaN values untouched so "
        "they still load as real SQL NULL rather than the string 'nan'.")
```

# Stage 2 — Conducting initial data investigation

```python
## @title 2.1 Inspect missing rows, then build the clean working dataframes
section_header("Cleaning missing review text", "SQLite: inspect what's being excluded, then query the clean subset")

# 1. See exactly what's being excluded, straight from the DB
google_missing = pd.read_sql(
    "SELECT * FROM google_reviews_raw WHERE Comment IS NULL", conn
)
trustpilot_missing = pd.read_sql(
    'SELECT * FROM trustpilot_reviews_raw WHERE "Review Content" IS NULL', conn
)

print(f"Google rows with missing Comment: {len(google_missing):,}")
print(f"Trustpilot rows with missing Review Content: {len(trustpilot_missing):,}")

display_table(google_missing[["Club's Name", "Creation Date", "Comment", "Overall Score"]].head(10),
              title="Google — sample rows with missing Comment (excluded)")
display_table(trustpilot_missing[["Location Name", "Review Created (UTC)", "Review Content", "Review Stars"]].head(10),
              title="Trustpilot — sample rows with missing Review Content (excluded)")

# 2. Query the CLEAN subset — this becomes the working dataframe for every later stage
google_df = pd.read_sql(
    "SELECT * FROM google_reviews_raw WHERE Comment IS NOT NULL", conn
).reset_index(drop=True)

trustpilot_df = pd.read_sql(
    'SELECT * FROM trustpilot_reviews_raw WHERE "Review Content" IS NOT NULL', conn
).reset_index(drop=True)

insight(f"Working dataframes: {len(google_df):,} Google rows, {len(trustpilot_df):,} Trustpilot rows. "
        f"9,352 of Google's 23,250 rows (40.2%) had no Comment text and were dropped — Trustpilot lost 0 of its "
        f"16,673 rows. This asymmetry is a genuine platform difference, not an error: Google lets users submit a "
        f"star rating alone with no comment box required, while Trustpilot's flow generally prompts for written "
        f"feedback. Practical implication: Google's word-level analysis draws on a smaller effective sample "
        f"(13,898) than Trustpilot's full 16,673 — worth keeping in mind when comparing results between "
        f"platforms later. Raw tables in the DB still have all {len(google_raw):,} / {len(trustpilot_raw):,} "
        f"rows if you ever need to go back and check what was excluded.")
```

```python
## @title 2.1b Check for other forms of "missing" text (whitespace, empty string, literal "nan")
section_header("Deeper quality check", "Catching blanks that aren't true NULLs")

def check_hidden_blanks(df, col, label):
    empty_string  = (df[col] == "").sum()
    whitespace    = df[col].astype(str).str.strip().eq("").sum()
    literal_nan   = df[col].astype(str).str.lower().eq("nan").sum()
    print(f"{label}: empty string={empty_string}, whitespace-only={whitespace}, literal 'nan' text={literal_nan}")
    return empty_string, whitespace, literal_nan

g_results = check_hidden_blanks(google_df, "Comment", "Google Comment")
t_results = check_hidden_blanks(trustpilot_df, "Review Content", "Trustpilot Review Content")

insight("Beyond true SQL NULLs, also checked for empty strings, whitespace-only entries, and stray literal 'nan' "
        "text (which can appear if a NaN gets cast to a string somewhere upstream). All three checks returned 0 "
        "for both Comment and Review Content — confirming google_df and trustpilot_df are genuinely clean, with "
        "no hidden blanks slipping through the NULL-based filter above.")
```

# Stage 3 — Conducting initial topic modelling

```python
## @title 3.1 Preprocessing — lowercase, remove stopwords & numbers
section_header("Preprocessing reviews", "Lowercase → remove numbers → remove stopwords → tokenise")

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"\d+", "", text)           # remove numbers
    text = re.sub(r"[^a-z\s]", " ", text)      # strip punctuation/symbols
    tokens = word_tokenize(text)
    tokens = [t for t in tokens if t not in STOP_WORDS and len(t) > 1]
    return tokens

google_df["tokens"]     = google_df["Comment"].apply(clean_text)
trustpilot_df["tokens"] = trustpilot_df["Review Content"].apply(clean_text)

google_df["clean_text"]     = google_df["tokens"].apply(" ".join)
trustpilot_df["clean_text"] = trustpilot_df["tokens"].apply(" ".join)

display_table(google_df[["Comment", "clean_text"]].head(), title="Google — before/after cleaning")
```

# Stage 5 — Conducting emotion analysis

```python
## @title 5.1 Load the emotion classification model
from transformers import pipeline

section_header("Emotion analysis", "bhadresh-savani/bert-base-uncased-emotion")

emotion_pipe = pipeline(
    "text-classification",
    model="bhadresh-savani/bert-base-uncased-emotion",
    top_k=None
)

example = "I am absolutely furious — they charged me a fee after I had already cancelled my membership."
print(emotion_pipe(example))
```

```python
## @title 5.2 Run emotion model on both negative-review sets (batched)
def top_emotions_batch(texts, batch_size=32):
    texts = [str(t)[:512] for t in texts]
    results = emotion_pipe(texts, batch_size=batch_size)
    return [max(r, key=lambda x: x["score"])["label"] for r in results]

google_neg["top_emotion"]     = top_emotions_batch(google_neg["Comment"].tolist())
trustpilot_neg["top_emotion"] = top_emotions_batch(trustpilot_neg["Review Content"].tolist())

display_table(google_neg[["Comment", "top_emotion"]].head(), title="Google negative reviews — top emotion")
```

# Stage 6 — Using a large language model from Hugging Face

Uses an instruction-tuned LLM (Falcon-7b-instruct, or Phi-4-mini via Ollama as a lighter alternative) purely
through prompting — no fine-tuning, consistent with Week 3's zero-shot prompting concept.

```python
## @title 6.1 Load the instruction-tuned LLM (Phi-4-mini-instruct, via transformers)
section_header("Loading the LLM", "microsoft/Phi-4-mini-instruct — switched from Falcon-7b (legacy trust_remote_code incompatible with transformers 5.x)")

from transformers import pipeline as hf_pipeline

llm = hf_pipeline(
    "text-generation",
    model="microsoft/Phi-4-mini-instruct",
    device_map="auto",
)

def llm_generate(prompt, max_tokens=80):
    messages = [{"role": "user", "content": prompt}]
    out = llm(messages, max_new_tokens=max_tokens, do_sample=False)
    return out[0]["generated_text"][-1]["content"]

print(llm_generate("Say hello in one short sentence.", 20))
```

```python
# @title 6.2 Prompt 1 — extract top 3 topics per review
section_header("LLM topic extraction", "microsoft/Phi-4-mini-instruct, zero-shot prompting")

def topic_prompt(review_text):
    return (f"In the following customer review, pick out the main 3 topics. "
            f"Return them in a numbered list format, with each one on a new line.\n\n"
            f"Review: \"{review_text}\"")

# Note: brief explicitly allows a subset if execution time is too high.
# Starting with 30 to confirm the pipeline works end-to-end before scaling up.
sample_reviews = angry_reviews[:30] if len(angry_reviews) > 30 else angry_reviews

extracted_topics = []
for review in sample_reviews:
    output = llm_generate(topic_prompt(review), max_tokens=80)
    extracted_topics.append(output)

print(extracted_topics[0])
```

```python
## @title 6.3 Re-run BERTopic on the LLM-extracted topic list
comprehensive_topic_list = extracted_topics  # one entry per review, each a short numbered-list string

topic_model_llm = BERTopic(language="english", verbose=True)
topics_llm, _ = topic_model_llm.fit_transform(comprehensive_topic_list)
display_table(topic_model_llm.get_topic_info(), title="Topics from LLM-extracted topic phrases")
```

```python
## @title 6.5 Prompt 2 — generate improvement suggestions (reduced scope for time)
def suggestion_prompt(topics_text):
    return (f"For the following text topics obtained from negative customer reviews, can you give some "
            f"actionable insights that would help this gym company?\n\nTopics:\n{topics_text}")

suggestions = []
for topics_text in comprehensive_topic_list[:10]:  # reduced from 30 to 10 given generation time
    output = llm_generate(suggestion_prompt(topics_text), max_tokens=80)  # reduced from 150 to 80
    suggestions.append(output)

print(suggestions[0])
```

# Stage 7 — Using Gensim (LDA)


```python
## @title 7.1 Preprocessing for Gensim LDA
section_header("Gensim LDA", "10 topics, combined negative reviews (Google + Trustpilot)")

from gensim import corpora
from gensim.models import LdaModel

lda_tokens = [toks for toks in (google_neg["tokens"].tolist() + trustpilot_neg["tokens"].tolist())]

dictionary = corpora.Dictionary(lda_tokens)
dictionary.filter_extremes(no_below=5, no_above=0.5)
corpus_bow = [dictionary.doc2bow(toks) for toks in lda_tokens]
```

```python
## @title 7.2 Train the LDA model (10 topics)
lda_model = LdaModel(corpus=corpus_bow, id2word=dictionary, num_topics=10,
                      random_state=42, passes=10, alpha="auto", eta="auto")

for i, topic in lda_model.print_topics(num_topics=10, num_words=8):
    print(f"Topic {i}: {topic}")
```
