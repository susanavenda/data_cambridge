# Demonstration 1.3.1 Exploring logistic regression

Follow the demonstration to learn how to employ logistic regression. In this demonstration, you'll learn to:
- implement a logistic regression model
- set a decision threshold
- deal with misclassification
- tune the model to improve accuracy.


### a. Import libraries and generate data.

```python
# Import required libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from ipywidgets import interactive, FloatSlider

import warnings
warnings.filterwarnings('ignore')
```

```python
# Generate data with some overlap.
np.random.seed(0)

X = np.random.normal(size=(100, 1))
bias = 0.5 # Adjust this bias to shift the decision boundary.
y = np.array([1 if x > (np.random.uniform(-0.4, 0.6) + bias) else 0 for x in X.ravel()])
```

### b. Create and fit a logistic regression model.

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
# Final Parameters, after the model has finished training
w = model.coef_[0][0]
b = model.intercept_[0]
```

```python
# Create a table of data with predictions.
X_values = X.flatten()

predictions_prob = model.predict_proba(X_values.reshape(-1, 1))[:, 1]
data = pd.DataFrame({'X': X_values, 'y': y, 'Predicted': predictions_prob})

# Style.
styled_data = data.style.set_table_styles(
    [{'selector': 'th',
        'props': [
            ('background-color', 'lightgrey'),
            ('font-weight', 'bold'),
            ('color', 'black'),
            ('text-align', 'center')
            ]
      }]
    ).set_properties(**{
        'text-align': 'center',
        'border-color': 'black',
        'border-style' :'solid',
        'border-width': '1px'
   })

styled_data
```

```python
def sigmoid_inverse(y):
    if 0 < y < 1:
        z = np.log(y / (1 - y))  # Find z
        return z
    else:
        raise ValueError("y must be between 0 and 1")
```

### c. Setting the threshold

```python
def plot_func(threshold_y):
    fig, ax = plt.subplots(figsize=(10, 5))
    colors = np.where(y == 0, 'skyblue', 'salmon')

    # Transform X to z as per the fitted model parameters
    z_data = w * X + b
    predicted_y = model.predict_proba(X)[:, 1]

    threshold_z = sigmoid_inverse(threshold_y)

    # Plot the original data points with larger circles
    ax.scatter(z_data[y == 0], y[y == 0], label='Class 0', alpha=0.7, color='blue', s=100, edgecolor='k')
    ax.scatter(z_data[y == 1], y[y == 1], label='Class 1', alpha=0.7, color='red', s=100, edgecolor='k')

    # Plot the predicted y values with high transparency
    ax.scatter(z_data, predicted_y, label='Predicted y (sigmoid output)', alpha=0.3, color='green')

    # Calculate accuracy.
    predictions = predicted_y > threshold_y
    accuracy = np.mean(predictions == y)

    # Draw the threshold lines (vertical).
    ax.axhline(y=threshold_y, color='red', linestyle='--')
    ax.axvline(x=threshold_z, color='green', linestyle='--')

    # Fill the areas.
    ax.axvspan(z_data.min(), threshold_z, color='skyblue', alpha=0.3)
    ax.axvspan(threshold_z, z_data.max(), color='salmon', alpha=0.3)

    ax.set_xlabel('z')
    ax.set_ylabel('Predicted y (Probability)')
    ax.legend()
    ax.set_title(f'Sigmoid Function with Threshold Line (Model Accuracy: {accuracy:.2f})')
    plt.show()


# Create interactive slider for y values.
threshold_slider = FloatSlider(value=0.5, min=0.001, max=0.999, step=0.01, description="Threshold:")
interactive_plot = interactive(plot_func, threshold_y=threshold_slider)

# Display the interactive plot.
interactive_plot
```

# Key information
This demonstration illustrated how to employ a logistic regression model and examine the effects of a specific threshold setting.

When selecting a threshold, we examined the classification outcomes for data points that had landed near the middle of the range. These were potential misclassifications.
