## Scenario 1: Customer churn prediction
A telecommunications company seeks to identify which customers are at high risk of churning (cancelling their service) in the near future based on their usage patterns, billing information, and customer service interactions.

**Approach: Classification**

Reason:
- The goal is to predict a binary outcome: churn (yes/no).
- The expected output is thus categorical (binary).
- Classification algorithms are designed to categorise data into discrete classes.

Implications:
- By identifying customers at high risk of churning, the company can take proactive measures (e.g. targeted promotions, personalised offers) to retain them.








## Scenario 2: Sales forecasting
A retail company wants to forecast next month's sales volume based on historical sales data, seasonality, and promotional activities.

**Approach: Regression**

Reason:

- The goal is to predict a continuous numeric value: sales volume.
- The expected output is a real number, making regression the suitable approach.

Implications:

- By predicting the upcoming month sales, the company can optimise inventory management, workforce planning, and budgeting.

## Scenario 3: Customer lifetime value prediction
A company wants to predict the lifetime value (LTV) of customers based on their purchasing history, engagement with marketing campaigns, and demographic data.

**Approach: Regression**

Reason:

- The goal is to predict a continuous numeric value: customer lifetime value (LTV).
- The expected output is a monetary estimate, making regression the appropriate approach.

Implications:

- By predicting customer LTV, the company can personalise customer engagement, allocate resources effectively, and maximise long-term profitability.

## Scenario 4: Credit score rating
A financial institution wants to predict the creditworthiness of its customers by assigning them to one of four categories (1, 2, 3, 4) based on their credit history, income level, and other financial indicators.

**Approach: Classification**

Reason:

- The goal is to assign customers to one of four discrete categories (1, 2, 3, 4).
- The expected output is categorical, making classification the appropriate approach.

Implications:
- Helps the financial institution assess risk, make informed lending decisions, and offer appropriate credit products to customers.

## Scenario 5: Predicting customer satisfaction levels
A company wants to predict customer satisfaction on a scale from 1 to 5 based on service usage data, customer feedback, and interaction history.

**Approach: Classification**

Reason:

- The goal is to predict customer satisfaction levels on a fixed ordinal scale (1 to 5).
- The expected output is categorical, making classification the appropriate approach.

Implications:

- Enables the company to identify areas for service improvement and enhance customer experience.

## Scenario 6: Age prediction
A social media company wants to predict the age group (e.g. 18–24, 25–34) of its users based on their activity, preferences, and interactions on the platform.

**Approach: Classification**

Reason:

- The goal is to predict the user's age group, which falls into discrete categories (e.g. 18–24, 25–34).
- The expected output is categorical, making classification the appropriate approach.

Implications:

- Helps in personalising content, targeting advertisements, and improving user engagement.
- Enhances demographic insights for better business and marketing strategies.
