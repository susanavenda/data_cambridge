```python
from activation_helper import tables
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Fabricate data
x = np.linspace(-10, 10, 100)
```

## b. Define the activation functions.

```python
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Softmax function
def softmax(x):
    e_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return e_x / e_x.sum(axis=1, keepdims=True)

# ReLU function
def relu(x):
    return np.maximum(0, x)
```

## c. Classify data.

```python
# Apply the functions
y_sigmoid = sigmoid(x)
y_relu = relu(x)

# Styling dataframes for better visual appeal
def highlight_max(s):
    '''Highlight the maximum in a series yellow.'''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

# Softmax sample data (multiple samples with multiple inputs)
softmax_samples = np.array([[-1, 0, 1], [2, 1, 0], [-1, -2, 3], [2, 5, -1], [0, -2, 2]])
y_softmax = softmax(softmax_samples)

# Select a few values for the table
selected_indices = np.linspace(0, len(x)-1, 5, dtype=int)
selected_x = x[selected_indices]
selected_y_sigmoid = y_sigmoid[selected_indices]
selected_y_relu = y_relu[selected_indices]
```

```python
tables(selected_x, selected_y_sigmoid, selected_y_relu, softmax_samples, y_softmax, highlight_max)
```
