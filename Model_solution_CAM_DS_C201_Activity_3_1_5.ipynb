# Model solution

## Building a basic neural network

```python
# Import the required libraries.
import pandas as pd
import numpy as np
```

## Read the data

```python
# Retrieve the URL to import data set from GitHub.
url = 'https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/spamdata.csv'
```

```python
# Import the data.
# Note that the data does not contain column names (headers).
spam = pd.read_csv('https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/spamdata.csv', header=None)

# View the data.
print(spam.shape)
spam.head()
```

## Create input features (X) and target variable (y)

```python
# Create a new variable X comprising of all the input features.
X = spam.drop(57, axis=1)

# Create a new variable y to store the target variable.
# Note Column #57 contains the target values.
y = spam.iloc[:, 57]
```

## Create a train-validation-test split

```python
# Import train_test_split.
from sklearn.model_selection import train_test_split

# Split the original data into train and test sets, with a 80:20 split.
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2)

# Split the train data further into train and validation sets, with a 90:10 split.
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,
                                                  test_size=0.1)
```

## Standardise the data

```python
# Standardise the features with the StandardScaler.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_val = scaler.transform(X_val)
```

## Define the sequential model

- 2 hidden layers, output layer
  - Input to Hidden Layer #1 will be the Input features matrix, X.
  - Input shape of Hidden Layer #1 = Number of columns/features of X.

- Hidden layer #1
  -  64 neurons with ReLU activition
- Hidden layer #2
  - 32 neurons with ReLU activation
- Output layer
    - 1 neuron with Sigmoid activation (for binary classification problem)
  

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

## Compile the model
- For a binary classification model, set the loss function as: binary_cross_entropy
- Track the accuracy metric.
- Use Adam Optimiser.

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

## Train the model
- Use batch size = 64.
- Set epochs = 10.
- Check model performance for each epoch on the validation data, as the model trains.

```python
batch_size = 64

epochs = 10

model.fit(X_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          validation_data=(X_val, y_val))
```

## Evaluate the trained model on the test data

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
