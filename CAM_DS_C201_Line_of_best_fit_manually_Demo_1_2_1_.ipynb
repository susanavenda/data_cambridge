# Demonstration 1.2.1 Finding the line of best fit manually

Follow the demonstration to see how to employ a supervised learning models on two data sets. In this demonstration, you’ll learn how:
* to construct a simple linear regression line
* to adjust the parameters manually to improve how the line fits the data.

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generate 100 linear data points with noise.
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = 2.5 * x + 3 + np.random.randn(100) * 2
```

```python
def plot_line(w, b):
    plt.scatter(x, y)
    y_pred = w * x + b
    plt.plot(x, y_pred, color='red')

    # Calculate R-squared and MSE.
    mse = mean_squared_error(y, y_pred)
    r_squared = r2_score(y, y_pred)

    plt.title(f"Linear model: y = {w:.2f}x + {b:.2f} | R^2: {r_squared:.3f} | MSE: {mse:.2f}")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim(y.min() - 5, y.max() + 5)
    plt.show()
```

```python
interactive_plot = interactive(plot_line, w=(-5.0, 5.0, 0.1), b=(-10, 10, 0.5))
interactive_plot
```
