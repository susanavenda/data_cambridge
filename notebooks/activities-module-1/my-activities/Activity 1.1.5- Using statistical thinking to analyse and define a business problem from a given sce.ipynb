# Activity 1.1.5: Using statistical thinking to analyse and define a business problem from a given scenario

## Scenario
**Communik8** is a telecommunications company with both online and physical stores. It provides mobile contracts and sells handsets. Recently, the company has experienced a sizeable reduction in new customers and a decrease in the retention of existing customers.

With the primary goal of improving business performance, the company needs to identify critical reasons for this recent decline. The company would like to make sense of its customer data and transform it into actionable insights.

The management team at Communik8 has provided recent customer feedback to your team of data scientists and would like initial feedback on the main issues that can be identified from this data.

## Objective
The goal is to explore and analyse an organisational scenario to identify key problems.

## Assessment criteria
By completing this activity, you will be able to provide evidence that you can:
* Note common themes and sentiments expressed by customers without performing any in-depth analysis.
* Formulate your own business problem statement.


## Activity guidance
1. Analyse the customer comments and categorise the comments into themes (e.g. customer service, delivery, billing and payment, and so on).
2. Using the themes you identified as guidance, define the problems Communik8's customers are experiencing. (e.g. delays in delivery, website functionality, and so on).
3. Identify how solving these problems will help the company achieve its objectives and goals, and outline the consequences of not achieving these goals (e.g. improved customer satisfaction, customer retention, loss of revenue, and so on).
4. Write a problem statement based on themes, problems, and solutions you identified.

# Model solution

### Analysing/categorising comments & defining problems

**Customer service**
- Wait time: 'CommuniK8 is the worst company ever – I waited 45 minutes for my call to be answered...'
- Response quality: '...and the customer service was useless!'
- Staff behaviour: 'The staff at the Riverside store are incredibly rude. I needed help with my handset, and they laughed at me.'
- Misleading information: 'I tried to cancel within the 14-day cooling-off period and was told it would happen automatically. It didn’t, and now I’m tied into an unwanted 24-month contract.'
- Delayed responses/unresolved complaints: 'I’m still waiting for a response to the complaint email I sent a month ago. Going to cancel, been a loyal customer for 10 years.'

**Website and digital experience**
- Website quality & functionality: 'The website is a joke, and the only way to ask questions is through the chat.'
- Chat quality: 'The responses are irrelevant and ungrammatical.'

**Delivery and fulfilment**
- Order fulfilment: 'My handset took two weeks to arrive, not the promised 48 hours, so I cancelled my order.'

**Network coverage**
- Service coverage: 'Coverage in my area is rubbish. I can’t make or receive calls, and the mobile internet connection is very patchy.'

**Billing and payments**
- Incorrect billing: 'I was charged twice in one month and got a debt collector’s letter when I refused to pay the bill.'
- Contract cancellation process: 'I tried to cancel within the 14-day cooling-off period and was told it would happen automatically. It didn’t, and now I’m tied into an unwanted 24-month contract.'




### Solving customer-service issues

**Wait time: Customers face excessively long wait times to have their calls answered**

* Impact: Customer frustration and dissatisfaction, leading to higher call
abandonment rates and potential loss of customers.
* Solution: Implement a callback system, increase staffing levels, and optimise call routing.

**Response quality: Inadequate and unhelpful customer service responses**

* Impact: Customers feel undervalued, leading to a decline in trust and brand reputation.
* Solution: Conduct customer service training to improve the support quality and ensure agents have access to comprehensive resources.

**Staff behaviour: Reports of unprofessional behaviour from in-store staff.**

* Impact: Poor brand perception, negative online reviews, and increased likelihood of customer churn.
* Solution: Implement staff training programs focused on empathy, communication, and conflict resolution.

**Misleading information: Customers are misinformed about the 14-day cooling-off period, resulting in unwanted contract commitments**

* Impact: Customer disputes, increased complaints, and potential legal action.
* Solution: Improve staff knowledge of policies and provide clear, accessible customer guidance through multiple channels (e.g. website, email, store signage).

**Unresolved complaints: Delays in complaint resolution, with some complaints going unanswered for over a month**

* Impact: Increased customer churn, reputational damage, and loss of loyal customers.
* Solution: Set response time targets for complaints, automate acknowledgement  emails, and create an escalation process for unresolved issues.

### Website and digital experience

**Website functionality: Customers face difficulties navigating the website, with limited options for direct inquiries**

* Impact: Increased call centre load, customer frustration, and negative user experience.
* Solution: Redesign the website to improve navigation, add a self-service help centre, and offer more communication options (e.g. email, FAQ).

**Chat quality: The chatbot provides irrelevant and ungrammatical responses.**

* Impact: Customer dissatisfaction and loss of trust in the digital support experience.
* Solution: Upgrade chatbot AI to provide context-aware, grammatically correct responses and ensure a seamless transfer to human support when needed.

### Delivery and fulfilment

**Order fulfilment: Delays in delivery, with promised 48-hour shipping turning into a two-week wait**

* Impact: Canceled orders, customer dissatisfaction, and loss of future sales.
* Solution: Improve supply chain processes, ensure accurate delivery timelines, and offer proactive customer updates on order status.

### Network coverage

**Service coverage: Reports of poor mobile coverage, resulting in dropped calls and patchy internet connections**

* Impact: Customer frustration, increased support calls, and churn to competitors.
* Solution: Conduct a coverage analysis, invest in network infrastructure, and provide location-based network status updates for customers.

### Billing and payments

**Incorrect billing: Double billing and debt collection issues**

* Impact: Customer disputes, risk of legal action, and brand damage.
* Solution: Implement billing audits, enhance system checks for billing errors, and give customers an easy way to dispute charges.

**Contract cancellation process: Customers face issues with the 14-day cooling-off period, leading to unwanted long-term contracts**

* Impact: Increase in formal complaints, regulatory scrutiny, and potential legal action.
* Solution: Provide transparent communication and automate cooling-off period cancellations.


### Writing a problem statement

**Problem statement**

CommuniK8 faces critical issues across multiple customer touchpoints, including prolonged wait times, poor customer service quality, rude staff behaviour, misleading contract information, unresolved complaints, website functionality issues, delivery delays, insufficient network coverage, and billing inaccuracies.

These challenges undermine customer satisfaction, retention, and brand loyalty, jeopardising long-term revenue and market competitiveness.

To address these problems, the company must prioritise reducing wait times, improving staff training, ensuring contract and billing processes transparency, upgrading digital platforms, enhancing delivery timelines, and expanding network coverage.

Failure to act will result in irreversible brand erosion, ongoing customer attrition, sustained revenue loss, deteriorating brand reputation, and a weakened competitive position in the telecommunications industry.
