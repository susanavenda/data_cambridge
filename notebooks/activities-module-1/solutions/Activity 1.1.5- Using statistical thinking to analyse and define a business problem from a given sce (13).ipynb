# Activity 1.1.5: Using statistical thinking to analyse and define a business problem from a given scenario

With the primary goal of improving business performance, the company needs to identify critical reasons for this recent decline. The company would like to make sense of its customer data and transform it into actionable insights.

The management team at Communik8 has provided recent customer feedback to your team of data scientists and would like initial feedback on the main issues that can be identified from this data.

## 1. Analysis of Themes and Sentiments

After reviewing the customer feedback summary, I have categorized the responses into the following **Nominal Categorical** themes:

* **Customer Service Inefficiency:** Reports of 45-minute wait times and "useless" or "irrelevant" automated chat responses.
* **Logistics & Fulfillment:** Handsets arriving in two weeks despite a 48-hour promise.
* **Billing & Contract Errors:** Systemic issues with double-charging and failures to process 14-day cooling-off cancellations.
* **Technical Performance:** Persistent "patchy" internet and poor mobile coverage.
* **Store Staff Conduct:** Negative interactions at physical locations (e.g., Riverside store).

**Overall Sentiment:** Highly negative. The recurring theme is a loss of trust due to broken service promises.

---

## 2. Problem Definition

The core issues Communik8 is facing involve a breakdown between business promises and operational execution:

1. **Support Scalability:** The current digital and phone support systems cannot handle the current **Volume** of queries.
2. **Process Veracity:** Billing and cancellation processes are unreliable, creating legal and reputational risks.
3. **Logistical Lag:** A failure to meet delivery timelines is driving immediate cancellations of new orders.

---

## 3. Impact and Consequences

* **Goals:** Resolving these friction points is essential to stabilizing customer retention and returning to a state of high **Data Maturity**.
* **Consequences:** Continued inaction will result in a sustained "sizeable reduction" in revenue, market share loss to competitors, and a permanent decline in brand equity.

---

## 4. Business Problem Statement

> Communik8 is experiencing a significant decline in customer retention and new acquisitions caused by systemic failures in support response times, billing accuracy, and handset delivery. These operational gaps are resulting in lost revenue and a negative brand reputation. The company must identify the root causes of these service delays and billing errors to implement a data-driven strategy that improves the customer experience and stabilizes the subscriber base.

---

## 5. Reflection and Key Learning

* **Challenges:** The primary challenge was using a small sample of qualitative feedback to infer broader systemic issues without falling into **statistical fallacies** or making assumptions based on intuition.
* **Key Learning:** I learned that effective problem analysis requires distinguishing between surface-level symptoms (like rude staff) and underlying causes (like poor system support or training).

**Process and Rationale:**
I applied **Statistical Thinking** to organize unstructured text into measurable themes. By utilizing the **5 Whys** approach, I moved from noting the "symptoms" of churn to identifying potential root causes in the company's technical and communication infrastructure. This ensures the project aligns with the **Business Understanding** phase of the **CRISP-DM** lifecycle.

1. Why has churn increased? There were numerous complaints about a new product.

2. Why were there complaints? The product had technical issues.

3. Why did it have technical issues? It was rushed through development to meet a deadline.

4. Why was it rushed? The project timeline wasn't adjusted despite early delays.

5. Why was the timeline not adjusted? Root Cause: There was a lack of communication between the development team and management.




# Reflect

To move the project into the Data Understanding and Modelling phases of the CRISP-DM lifecycle, I propose the following research question:

"How can we reduce the customer churn rate by 10% within the next six months by improving the response accuracy of digital support systems and resolving 95% of billing disputes within 48 hours?"

S (Specific): Targets digital support accuracy and billing dispute resolution.

M (Measurable): Sets a target of 10% churn reduction and 95% resolution rate.

A (Achievable): Focuses on internal process improvements using existing customer feedback data.

R (Relevant): Directly addresses the decline in retention and business performance goals.

T (Time-bound): Establishes a clear six-month timeframe for results.

By converting a vague problem into a SMART research question, the project becomes measurable and focused on a specific timeframe, significantly increasing the likelihood of delivering actionable insights to stakeholders
