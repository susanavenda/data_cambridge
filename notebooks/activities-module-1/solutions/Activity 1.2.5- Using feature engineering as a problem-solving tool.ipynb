# Model solution

**Cart abandonment rate:**
Customers abandon their carts if the website experience is slow, confusing, or frustrating. This is directly related to issues regarding website quality and chat support. We can identify friction points in the customer journey by monitoring the abandonment rate.

**Customer support response time:**
Response time directly reflects how long customers wait to receive support. If response times are long, it signals an operational bottleneck that can frustrate customers. Customers forced to wait longer for assistance are more likely to abandon calls, file complaints, or churn.By tracking response times, we can identify when and where delays occur, enabling process improvements such as staffing adjustments or automation.

**Conversion rate:**
Low conversion rates may signal that customers are visiting the site but not finding what they need, encountering usability issues, or facing unclear value propositions. Identifying the factors driving low conversion rates, such as pricing, product information, or checkout usability, can improve customer experience and help increase sales.

**Customer satisfaction score (CSAT):**
Direct feedback from customers provides qualitative insights into their experience with products, services, and support. CSAT also captures the customer’s emotional response, which strongly correlates with their likelihood of making repeat purchases or recommending the brand.

**Repeat purchase rate:**
This metric indicates customer loyalty and satisfaction. A low rate might reveal issues with product quality, post-purchase support, or customer experience. Tracking this rate helps identify how well customer needs are met and if there's an opportunity to improve customer retention.

**Website traffic:**
website traffic is a useful metric but will need to be used in conjunction with other metrics to enhance its effectiveness.
- Low traffic + Low conversion rate → Issues with marketing, awareness, or product appeal
- High traffic + Low conversion rate → Issues with user experience (UX), pricing, or checkout flow

Analysing traffic paths (e.g. exit pages) highlights points where potential customers leave the website. This may indicate content relevance, page load speed, or navigation issues. If traffic suddenly drops, it could indicate technical issues like site outages or slow load times, which can frustrate customers and cause abandonment.
