# Reflect

### 1. Selected Features for Analysis

To address the decline in performance at Communik8, I have prioritized the following **seven features** that bridge the gap between operational data and customer sentiment:

| Feature | Rationale for Selection |
| --- | --- |
| **Customer Support Response Time** | High wait times (45+ mins) were a top complaint. This is a primary driver of **immediate churn**. |
| **Average Shipping Time** | Measures the "broken promise" gap between the 48-hour guarantee and the reported 2-week reality. |
| **Repeat Purchase Rate** | The ultimate indicator of brand health. A drop here signals that existing customers no longer see value. |
| **Customer Satisfaction (CSAT) Score** | Quantifies the qualitative anger found in feedback, allowing us to correlate "unhappiness" with "cancellation." |
| **Cart Abandonment Rate** | Diagnoses the "New Customer" problem. High rates suggest friction in the digital sales funnel or hidden costs. |
| **Loyalty Membership** | Allows us to segment the data. Losing "Gold" members is a much more severe systemic threat than losing "Trial" users. |
| **Purchase Amount** | Identifies the **Financial Velocity**. It helps us see if we are losing high-value contract holders or low-tier prepaid users. |

---

### 2. Reflection: Process and Rationale

**Process:** My approach was guided by the **CRISP-DM** "Data Understanding" phase. I visualized the customer lifecycleâ€”from acquisition (Cart Abandonment) to post-purchase support (Response Time). I deliberately excluded internal financial metrics like "Inventory Turnover" because, while important for accounting, they do not explain the **psychology of churn**.

**Rationale:** By focusing on features that represent **Velocity** (speed of service) and **Veracity** (truth of the billing/experience), I am ensuring the model addresses the *root causes* of the business problem rather than just the symptoms. This ensures the final insights are actionable for the Head of Sales.

---

### 3. Key Learning

**In one sentence:** I learned that effective feature engineering is less about the quantity of data and more about selecting variables that serve as "proxies" for the real-world human experience and pain points.
